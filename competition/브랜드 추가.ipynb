```python
items = performance['브랜드']
encoder2= LabelEncoder()
encoder2.fit(items)

labels2 = performance['브랜드'] = encoder2.transform(items)
print('인코딩 변환값:',labels2)
```

```python
realname2 = []
encodedname2=[]
for i, item in enumerate(encoder2.classes_) : 
    print(item, '->', i)
    realname2.append(item)
    encodedname2.append(i)
```

```python

brand_name_encode = pd.DataFrame()
brand_name_encode['name']=realname2
brand_name_encode['encode_name']=encodedname2

brand_name_encode
```
